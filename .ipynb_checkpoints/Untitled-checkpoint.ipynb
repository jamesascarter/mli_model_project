```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# Check for GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define transformations (convert images to tensors and normalize)
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

# Load MNIST dataset
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, transform=transform, download=True)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, transform=transform, download=True)

from collections import Counter

# Get all labels in training dataset
train_labels = [label for _, label in train_dataset]

# Count occurrences of each digit
train_counts = Counter(train_labels)

# Print distribution
print("Training Data Distribution:")
for digit, count in sorted(train_counts.items()):
    print(f"Digit {digit}: {count} samples")

# Visualize with a bar chart
import matplotlib.pyplot as plt

plt.bar(train_counts.keys(), train_counts.values(), color='blue')
plt.xlabel("Digit")
plt.ylabel("Number of Samples")
plt.title("Training Data Distribution")
plt.show()

# Get all labels in test dataset
test_labels = [label for _, label in test_dataset]

# Count occurrences
test_counts = Counter(test_labels)

# Print distribution
print("\nTest Data Distribution:")
for digit, count in sorted(test_counts.items()):
    print(f"Digit {digit}: {count} samples")

# Visualize with a bar chart
plt.bar(test_counts.keys(), test_counts.values(), color='red')
plt.xlabel("Digit")
plt.ylabel("Number of Samples")
plt.title("Test Data Distribution")
plt.show()


# Create DataLoader for batch processing
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Define a simple neural network
class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(28*28, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 10)  # 10 classes (digits 0-9)

    def forward(self, x):
        x = x.view(-1, 28*28)  # Flatten image into 1D vector
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# Initialize model, loss function, and optimizer
model = NeuralNet().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 5
for epoch in range(num_epochs):
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# Evaluate on test data
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Test Accuracy: {100 * correct / total:.2f}%")

# Display some test images with predictions
import numpy as np

images, labels = next(iter(test_loader))
images, labels = images.to(device), labels.to(device)
outputs = model(images)
_, predicted = torch.max(outputs, 1)

fig, axes = plt.subplots(1, 5, figsize=(10, 3))
for i, ax in enumerate(axes):
    ax.imshow(images[i].cpu().numpy().squeeze(), cmap='gray')
    ax.set_title(f"Pred: {predicted[i].item()}")
    ax.axis('off')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
from torchvision.transforms import RandomRotation, RandomAffine

# Check for GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define standard transforms
base_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Define targeted augmentation only for 3 and 5
class AugmentedMNIST(Dataset):
    def __init__(self, original_dataset, augment_labels=[3, 5]):
        self.dataset = original_dataset
        self.augment_labels = augment_labels
        self.base_transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))
        ])
        self.augment_transform = transforms.Compose([
            RandomRotation(degrees=15),
            RandomAffine(degrees=0, translate=(0.1, 0.1)),
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        image, label = self.dataset[idx]

        # Apply augmentation only for digits 3 and 5
        if label in self.augment_labels:
            image = self.augment_transform(image)
        else:
            image = self.base_transform(image)

        return image, label

# Load MNIST dataset with targeted augmentation
original_train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True)
original_test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True)

train_dataset = AugmentedMNIST(original_train_dataset)
test_dataset = AugmentedMNIST(original_test_dataset)
# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Define an improved neural network
class ImprovedNeuralNet(nn.Module):
    def __init__(self):
        super(ImprovedNeuralNet, self).__init__()
        self.fc1 = nn.Linear(28*28, 256)  # More neurons
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(256, 128)  
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(128, 10)

    def forward(self, x):
        x = x.view(-1, 28*28)  # Flatten image
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        return x

# Initialize model, loss function, and optimizer
model = ImprovedNeuralNet().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 5
for epoch in range(num_epochs):
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# Evaluate on test data
model.eval()
correct = 0
total = 0
y_true = []
y_pred = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        
        # Collect predictions for confusion matrix
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

print(f"Test Accuracy: {100 * correct / total:.2f}%")

# Plot confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=range(10), yticklabels=range(10))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Display some test images with predictions
images, labels = next(iter(test_loader))
images, labels = images.to(device), labels.to(device)
outputs = model(images)
_, predicted = torch.max(outputs, 1)

fig, axes = plt.subplots(1, 5, figsize=(10, 3))
for i, ax in enumerate(axes):
    ax.imshow(images[i].cpu().numpy().squeeze(), cmap='gray')
    ax.set_title(f"Pred: {predicted[i].item()}")
    ax.axis('off')
plt.show()
```
